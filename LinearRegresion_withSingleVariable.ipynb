$$J(\theta) = \frac{1}{2m} \sum_{i=0}^m (h_\theta(x_i) - y_i)^2$$

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=100, n_features=1, noise=7, random_state=21)
```

```python
plt.scatter(x, y, color='red', label='Features')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x, y)
```

```python
lr.coef_
```

```python
p = lr.predict(x)
```

```python
plt.scatter(x, y, color='blue')
plt.plot(x, p, color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
XTrain, XTest, YTrain, YTest = train_test_split(x, y, test_size=0.25, random_state=2)
```

```python
plt.scatter(XTrain, YTrain)
plt.scatter(XTest, YTest, color='red')
plt.show()
```

```python
lr.fit(XTrain, YTrain)
```

```python
p = lr.predict(XTest)
```

```python
plt.scatter(XTest, YTest)
plt.plot(XTest, p, color='red')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(YTest, p)
```

```python
q = lr.predict(XTrain)
```

```python
mean_squared_error(YTrain, q)
```

```python
x, y = make_regression(n_samples=10, n_features=1, noise=2.5, random_state=1)
plt.scatter(x, y, color='red', label='Features')
plt.show()
```

```python
XTrain, XTest, YTrain, YTest = train_test_split(x, y, test_size=0.2, random_state=2)
plt.scatter(XTrain, YTrain)
plt.scatter(XTest, YTest, color='red')
plt.show()
```

```python
lr.fit(XTrain, YTrain)
p = lr.predict(XTest)
plt.scatter(XTest, YTest)
plt.plot(XTest, p, color='red')
plt.show()
```

```python
q = lr.predict(XTrain)
mean_squared_error(YTest, p), mean_squared_error(YTrain, q)
```
